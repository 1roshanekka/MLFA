```python
import pandas as pd
import seaborn as sns
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
df=pd.read_csv("soil.csv")
df=df.dropna()
df2=df.sample(n=155)
```

```python
clustering_model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='single')
clustering_model.fit(df2.iloc[:,1:6])
clustering_model.labels_
data_labels = clustering_model.labels_
sns.scatterplot(x='B %', 
                y='S %', 
                data=df2, 
                hue=data_labels,
                palette="rainbow")
```

```python
clustering_model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
clustering_model.fit(df2.iloc[:,1:6])
clustering_model.labels_
data_labels = clustering_model.labels_
sns.scatterplot(x='B %', 
                y='S %', 
                data=df2, 
                hue=data_labels,
                palette="rainbow")
```

```python
clustering_model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
clustering_model.fit(df2.iloc[:,1:6])
clustering_model.labels_
data_labels = clustering_model.labels_
sns.scatterplot(x='B %', 
                y='S %', 
                data=df2, 
                hue=data_labels,
                palette="rainbow")
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")

clusters = shc.linkage(df2.iloc[:,1:6], 
            method='average', 
            metric="euclidean")
shc.dendrogram(clusters)
plt.axhline(y = 85, color = 'r', linestyle = '-')
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")

clusters = shc.linkage(df2.iloc[:,1:6], 
            method='single', 
            metric="euclidean")
shc.dendrogram(clusters)
plt.axhline(y = 40, color = 'r', linestyle = '-')
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")

clusters = shc.linkage(df2.iloc[:,1:6], 
            method='complete', 
            metric="euclidean")
shc.dendrogram(clusters)
plt.axhline(y = 120, color = 'r', linestyle = '-')
```
